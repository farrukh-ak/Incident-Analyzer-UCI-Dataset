```python
import sqlite3
import pandas as pd

# Load the cleaned incident_final CSV (export from preprocessing)
incident_final = pd.read_csv("outputs\incident_cleaned.csv", parse_dates=["opened_at", "resolved_at", "closed_at"])

# Connect / create SQLite database
conn = sqlite3.connect("incident_intelligence.db")

# Load table into SQLite
incident_final.to_sql("incidents_cleaned", conn, if_exists="replace", index=False)

print("Data loaded into SQLite successfully")
conn.close()
```
